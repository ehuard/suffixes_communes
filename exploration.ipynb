```python
import re

def custom_split(input_string):
    input_string = input_string.lower()
    # Split by either space or hyphen
    return re.split(r' |-|\'', input_string)

name = "Cormoranche-sur-Saône l'Ain"

custom_split(name)
```

```python
dico = {}

def add2dico(suffixe):
    if suffixe in dico:
        dico[suffixe]+=1
    else:
        dico[suffixe]=1

for name in all_names:
    for part in custom_split(name):
        if len(part)>2:
            add2dico(part[-2:])
            if len(part)>3:
                add2dico(part[-3:])
                if len(part)>4:
                    add2dico(part[-4:])
                    if len(part)>5:
                        add2dico(part[-5:])
```

```python
sorted_dico = sorted(dico.items(), key=lambda x:-x[1])
print(sorted_dico)
```

```python
def extract_substrings(input_string, substring_length):
    substrings = []
    for i in range(len(input_string) - substring_length + 1):
        substrings.append(input_string[i:i+substring_length])
    return substrings

suff = "anche"
taille = len(suff)
for b in range(2,taille):
    print(extract_substrings(suff, b))
```

```python
# Nettoyage du dictionnaire: si un suffixe de k lettres est présent plus de 75 fois, 
# on enlève les suffixes de moins de k lettres qui le comprennent
dico2 = dict(dico)
for suffixe, occ in dico.items():
    if occ > 75: # si beaucoup d'occurences, on enleve les versions plus petites de ce meme suffixe
        length = len(suffixe)
        for sub_len in range(2, length):
            for sub in extract_substrings(suffixe, sub_len):
                if sub in dico2:
                    del dico2[sub]
    else : # si peu d'occurences, on rejette
        if suffixe in dico2:
            del dico2[suffixe]
```

```python
sorted_dico = sorted(dico2.items(), key=lambda x:-x[1])
len(sorted_dico)
```
